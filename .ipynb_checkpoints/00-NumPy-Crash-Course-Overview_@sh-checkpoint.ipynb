## Indexing

```python
mat = np.arange(0,100).reshape(10,10)
```

```python
mat
```

```python
row = 0
col = 1
```

```python
mat[row,col]
```

```python
# With Slices
mat[:,col]
```

```python
mat[row,:]
```

```python
mat[0:3,0:3]
```

## Masking

```python
mat > 50
```

```python
mat[mat>50]
```
